```python
import pandas as pd

import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Define the tickers
tickers = ["^GSPC", "AAPL", "GOOG", "MSFT", "AMZN"] 
print(tickers)
print(tickers, flush=True)
```

```python
# Define the date range
start_date = "2014-01-01"
end_date = "2024-12-31"
print(start_date)
print(start_date, flush=True)
print(end_date)
print(end_date, flush=True)
```

```python
# Download the data
data = yf.download(tickers, start=start_date, end=end_date)

# Extract the adjusted close prices
adj_close = data["Adj Close"]

# Calculate cumulative returns
cumulative_returns = (adj_close / adj_close.iloc[0]) * 100
print(data.head(), flush=True)
print(cumulative_returns.head(), flush=True)
```

```python
# Plot the cumulative returns
plt.figure(figsize=(12, 6))
for ticker in cumulative_returns.columns:
  plt.plot(cumulative_returns[ticker], label=ticker) 

plt.title("Cumulative Returns (2014-2024)")
plt.xlabel("Date")
plt.ylabel("Cumulative Return (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Further analysis (examples)

# Calculate annualized returns
annualized_returns = (cumulative_returns.iloc[-1] / 100) ** (1/10) - 1 
print("Annualized Returns:")
print(annualized_returns)
```

```python
# Measure volatility (standard deviation of daily returns)
daily_returns = adj_close.pct_change()
volatility = daily_returns.std() * (252 ** 0.5)  # Annualized volatility
print("\nVolatility:")
print(volatility)
```

```python
# Correlation analysis
correlation_matrix = daily_returns.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# --- Calculate dollar gains ---
initial_prices = adj_close.iloc[0]  # Get the initial prices
dollar_gains = adj_close - initial_prices  # Calculate the difference

# --- Print dollar gains ---
print("\nDollar Gains:")
print(dollar_gains.head(), flush=True)
```

```python
# --- Plot the dollar gains ---
plt.figure(figsize=(12, 6))
for ticker in dollar_gains.columns:
  plt.plot(dollar_gains[ticker], label=ticker)

plt.title("Dollar Gains (2014-2024)")
plt.xlabel("Date")
plt.ylabel("Dollar Gain")
plt.legend()
plt.grid(True)
plt.show()
```

```python
with pd.ExcelWriter('stock_data.xlsx') as writer:
    adj_close.to_excel(writer, sheet_name='Adjusted_Close')
    dollar_gains.to_excel(writer, sheet_name='Dollar_Gains')
    cumulative_returns.to_excel(writer, sheet_name='Cumulative_Returns')
```
